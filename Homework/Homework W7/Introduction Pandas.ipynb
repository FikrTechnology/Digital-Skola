## import Library pandas

```python
import pandas as pd
import numpy as np
```

# Create Dataframe

### Create from List

```python
list_1 = [1,2,3,4,5,6]
list_2 = ['Muhammad', 'Fikri', 'Lala', 'Yeye', 'lali', 'als', 'uu']
column_names = ['Number', 'Name']

df = pd.DataFrame(list(zip(list_1, list_2)), columns=column_names)

print(df)
```

### Create from Array

```python
no_induk = [100, 200, 300]
nama = ['A', 'B', 'C']
nilai =[9.4, 7.6, 10]

siswa = np.array([no_induk, nama, nilai])

print(siswa)
```

```python
df2 = pd.DataFrame(siswa)
df2
```

### Create from Series

```python
no_induk = pd.Series(no_induk)
nama = pd.Series(nama)
nilai = pd.Series(nilai)

df3 = pd.DataFrame(no_induk, columns=['No Induk'])
df3['Nama'] = nama
df3['Nilai'] = nilai

df3
```

### Create from Tuple

```python
numbers = tuple([1,2,3,4])
alphabets = tuple(['A', 'B', 'C', 'D'])

df4 = pd.DataFrame([numbers, alphabets]).T
df4
```

### Create Dictionary

```python
kota = ['Jakarta', 'Medan', 'Banyuwangi']
jml_pdd = [15000000, 7000000, 1500000]

dict_kota = {'kota':kota,
             'Penduduk':jml_pdd}

df5 = pd.DataFrame(dict_kota)
df5
```
