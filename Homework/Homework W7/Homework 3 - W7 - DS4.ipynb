# DataFrame yang ke-1 Daftar Mahasiswa

Dataframe pertama isinya mengenai nilai dari mahasiswa jurusan statistika. Di dalam dataframe tersebut harus meliputi nama, id mahasiswa, mata kuliah, dan nilainya. Dataframe HARUS dibuat dengan menggunakan dictionary.

```python
nama = ['Muhammad Fikrie', 'Natalia Ringo', 'Satria Triputra', 'Tita Alfiani', 'Zazkia Dwi Tristia']
id = [101, 201, 102, 103, 301]
makul = ['Statistika', 'Sistem Informasi', 'Statistika', 'Statistika', 'Multimedia']
nilai = ['90', '88', '83', '97', '86']

dict_siswa = {
    'ID Siswa':id,
    'Nama Siswa':nama,
    'Mata Kuliah':makul,
    'Nilai':nilai
}
df = pd.DataFrame(dict_siswa)
df
```

```python
df.loc[(df['Mata Kuliah'] == 'Statistika') & (df['Nilai'] > '80')]
```

# DataFrame yang ke-2 Daftar Nama Hewan

Dataframe kedua isinya adalah kumpulan informasi hewan di kebun binatang. Informasi yang diperlukan adalah jenis hewan, nama panggilan, kapan dilahirkan, usia, dan jenis kelamin. Dataframe HARUS dibuat dengan menggunakan tuple.

```python
hewan = tuple(['Jerapah Africa', 'Gajah Sumatra', 'Harimau Sumatra', 'Kanguru Australia', 'Koala Australia'])
n_panggilan = tuple(['Jerapah', 'Gajah', 'Harimau', 'Kanguru', 'Koala'])
lahir = tuple(['11-08-2015', '12-05-2018', '27-08-2012', '21-01-2014', '13-03-2017'])
usia = tuple([5, 3, 9, 7, 4])
gender = tuple(['Jantan', 'Jantan', 'Betina', 'Jantan', 'Betina'])

df2 = pd.DataFrame([hewan, n_panggilan, lahir, usia, gender], index=['Nama', 'Nama Panggilan', 'Kelahiran', 'Usia', 'Gender']).T
df2
```

```python
df2.loc[(df2['Usia'] >= 5) & (df2['Gender'] == 'Jantan')]
```
