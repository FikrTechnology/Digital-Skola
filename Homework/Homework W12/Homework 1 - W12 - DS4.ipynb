```python
import pandas as pd
import string
import nltk
from nltk.tokenize import word_tokenize
import re
```

```python
df = pd.read_csv('/content/sample_data/Homework - Advance Data Preprocessing.csv')
df
```

```python
def cleaning_tokenizing(txt):
  txt = txt.lower() # menjadi huruf kecil semua
  txt = txt.translate(str.maketrans("","",string.punctuation)) #hapus symbol
  txt = nltk.tokenize.word_tokenize(txt) # tokenize
  return txt
df["texts"].apply(cleaning_tokenizing)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
text_counts= cv.fit_transform(df["texts"])
pd.DataFrame(text_counts.toarray(), columns = cv.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
text_counts= cv.fit_transform(df["texts"])
pd.DataFrame(text_counts.toarray(), columns = cv.get_feature_names())
```
