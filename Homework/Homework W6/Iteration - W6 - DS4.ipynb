```python
start = 10
sum_number = 0

while start < 20:
  kuadrat = start ** 2
  start += 1
  sum_number = sum_number + kuadrat
print('Jumlah kuadrat : {}'.format(sum_number))
```

```python
angka = int(input('silahkan masukkan angka: '))
bar = 1

while bar < angka + 1:
  bar_1 = 1
  while bar_1 < bar + 1:
    print(bar_1, end=' ')
    bar_1 += 1
  print('\n')
  bar += 1
```
