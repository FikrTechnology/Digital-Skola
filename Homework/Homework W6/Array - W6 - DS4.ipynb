```python
zona_merah_tuples = [('Jawa Tengah', 5135, 5318), ('Banten', 883, 954), ('Kalimantan Selatan', 969, 1033), ('Sulawesi Selatan', 1630, 1668)]

for i in zona_merah_tuples:
  precentage = ((i[2] - i[1])/ i[1])
  print('{0} : {1}'.format(i[0], precentage))
```
