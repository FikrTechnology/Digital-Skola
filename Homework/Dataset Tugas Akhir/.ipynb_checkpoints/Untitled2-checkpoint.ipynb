# Missing Value

```python
total = df.isnull().sum().sort_values(ascending=False)
percent1 = df.isnull().sum() / df.isnull().count() * 100
percent2 = percent1.sort_values(ascending=False)
missing_data = pd.concat([total, percent2], axis=1, keys=['Total', '%'])
missing_data
```

```python
# melihat isi value yang unik & value terbanyak

sunshine = df['Sunshine'].unique()
sunshine = np.sort(sunshine)
print(sunshine)
print(df['Sunshine'].mode())
```

```python
evaporation = df['Evaporation'].unique()
evaporation = np.sort(evaporation)
print(evaporation)
print(df['Evaporation'].mode())
```

```python
# Values terbanyak di dalam kolom
print(df['Cloud3pm'].mode())
print(df['Cloud9am'].mode())
```

```python
# Menghapus kolom dengan Missing Value di atas 35%
df.drop(columns=['Sunshine', 'Evaporation', 'Cloud3pm', 'Cloud9am'], axis=1, inplace=True)
```

```python
df
```

```python
# Menghapus missing value di kolom Rain Today dan Rain Tomorrow
df.dropna(subset=['RainToday', 'RainTomorrow'], inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
df.describe(include=[object])
```

```python
# Imputasi Numeric

# handling missing value di kolom Pressure 9am&3pm di ubah menggunakan method mode()
df['Pressure9am'].fillna(df['Pressure9am'].mode, inplace=True)
df['Pressure3pm'].fillna(df['Pressure3pm'].mode, inplace=True)

# handling missing value di kolom Temp 9am&3pm di ubah menggunakan method mode()
df['Temp9am'].fillna(df['Temp9am'].mode, inplace=True)
df['Temp3pm'].fillna(df['Temp3pm'].mode, inplace=True)

# handling missing value di kolom Humidity 9am&3pm di ubah menggunakan method mode()
df['Humidity9am'].fillna(df['Humidity9am'].mode, inplace=True)
df['Humidity3pm'].fillna(df['Humidity3pm'].mode, inplace=True)

# handling missing value di kolom WindSpeed 9am&3pm di ubah menggunakan method mode()
df['WindSpeed9am'].fillna(df['WindSpeed9am'].mode, inplace=True)
df['WindSpeed3pm'].fillna(df['WindSpeed3pm'].mode, inplace=True)
df['WindGustSpeed'].fillna(df['WindGustSpeed'].mode, inplace=True)

# handling missing value di kolom Min MaxTemp di ubah menggunakan method mode()
df['MinTemp'].fillna(df['MinTemp'].mode, inplace=True)
df['MaxTemp'].fillna(df['MaxTemp'].mode, inplace=True)
```

```python
# Imputasi Categorical

# Handling missing value WindGustDir menggunakan method mode()
df['WindGustDir'].fillna(df['WindGustDir'].mode, inplace=True)

# Handling missing value WindDir 9am&3pm menggunakan method mode()
df['WindDir9am'].fillna(df['WindDir9am'].mode, inplace=True)
df['WindDir3pm'].fillna(df['WindDir3pm'].mode, inplace=True)
```
