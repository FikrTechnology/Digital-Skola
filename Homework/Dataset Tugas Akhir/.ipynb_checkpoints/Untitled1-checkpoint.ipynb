```python
# Handling missing value rainfall

df['Rainfall'] = df['Rainfall'].fillna(df['Rainfall'].mean())
```

```python
x = df[['Date', 'Rainfall']]
x['Year'] = x['Date'].apply(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d').strftime('%Y'))
x['Month'] = x['Date'].apply(lambda x : datetime.datetime.strptime(x, '%Y-%m-%d').strftime('%Y-%m'))
```

```python
# grafik rata" curah hujan selama 10 th.
plt.figure(figsize=(20,10))
x.groupby('Year')['Rainfall'].mean().plot(grid=True, marker='o')
plt.show()
```

```python
# grafik rata" curah hujan perbulan selama 10 th
plt.figure(figsize=(20,10))
x.groupby('Month')['Rainfall'].mean().plot(grid=True, marker='o')
plt.show()
```

```python
sns.lineplot(data=x, x="Year", y='Rainfall')
```

```python
# menampilkan urutan rata" curah hujan tertinggi di berbagai wilayah.
plt.figure(figsize=(20,15))
df.groupby('Location')['Rainfall'].mean().sort_values().plot(kind='barh', grid=True)
plt.show()
```

```python
x = df.groupby('Location')['Rainfall'].mean().sort_values()
plt.figure(figsize=(20,15))
x.tail().plot(kind='barh', grid=True)
plt.xlabel('Rainfall')
plt.show()
```

```python
# Contoh Persentase rata" curah hujan tertinggi di beberapa Lokasi
plt.figure(figsize=(20, 12))
df.groupby('Location')['Rainfall'].mean().sort_values().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
# menampilkan distribusi dari kolom Rain Tomorrow menggunakan hist
df['RainTomorrow'].hist()
```

```python
# Menghapus missing value dari kolom Rain Tomorrow dan Rain Today

df.dropna(subset=['RainTomorrow', 'RainToday'], inplace=True)
df.isnull().sum()
```

```python
# me Replace kolom Rain Today dan Rain Tomorrow menjadi numerik 0 dan 1

df[['RainToday', 'RainTomorrow']] = df[['RainToday', 'RainTomorrow']].replace({'Yes':1, 'No':0})
```
