#Question (include script you use to get the answer) :
1. What is the comparison between passengers that Survived and Not Survived ?
2. How many Survived passengers from each pclass ?
3. What is the average and median age of passengers that Survived and Not Survived ?
4. Create a new column named “age_interval” with this rules :
 Age < 25 : “< 25”
Age from 25 - 35 : “25 - 35”
Age from 36 - 50 : “36 - 50”
Age > 50 : “ > 50”
5. After creating the new column in question number 4, which age interval has the highest percentage of Survived ? (clue : percentage of survived is total survived per age interval /total all passengers in that age interval

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/sample_data/train.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# Jawaban Nomor 1

selamat = df['Survived'].sum()
tidak_selamat = df['Survived'].count() - selamat

print('Penumpang yang selamat adalah {} jiwa'.format(selamat))
print('Penumpang yang tidak selamat adalah {} jiwa'.format(tidak_selamat))
```

```python
# Jawaban Nomor 2

df.groupby('Pclass').agg(Survived = ('Survived','sum'))
```

```python
# Jawaban Nomor 3

df.groupby('Survived').agg(Rata_rata_Usia = ('Age','mean'),
                           Median_Usia = ('Age','median'),
                           )
```

```python
# Jawaban Nomor 4

df.loc[(df['Age'] < 25), 'Age_Interval'] = '<25'
df.loc[((df['Age'] > 25) | (df['Age'] == 35)), 'Age_Interval'] = '25 - 35'
df.loc[((df['Age'] > 35) | (df['Age'] == 50)), 'Age_Interval'] = '36 - 50'
df.loc[(df['Age'] > 50), 'Age_Interval'] = '> 50'
df
```

```python
# Jawaban Nomor 5

selamat = df.groupby('Age_Interval').agg(Total_Selamat = ('Survived','sum'),
                                         Total_penumpang = ('Age','count'))

selamat['Persentase'] = (selamat['Total_Selamat'] / selamat['Total_penumpang'].sum()) * 100
selamat
```

```python
test = (df.groupby('Age_Interval')['Survived'].sum() / df.groupby('Age_Interval')['Age'].count()) * 100

test
```
