<h2>Soal 1: Comparison Operator</h2>

- Berikan contoh comparison dari tipe data string 
- Berikan contoh comparison dari tipe data boolean
- Berikan contoh comparison dari tipe data integer

```python
#contoh comparison operator dari tipe data string

print('Fikri' == 'fikri')
print('Fikri' != 'fikri')
```

```python
#contoh comparison operator dari tipe data boolean

print(True == True)
print(True == False)
print(True != True)
print(True != False)
```

```python
#contoh comparison operator dari tipe data integer

print(10 > 4)
print(10 < 5)
print(10 <= 10)
print(11 >= 9)
print(10 == 10)
print(10 != 10)
```

<h2>Soal 3: If-Else Statement</h2>

Lengkapi kode untuk menghasilkan suatu output yang di harapkan

- Bualah sebuah if-else statement yang dimana akan mem-print 'Besar' jika ruangan adalah 'Kamar' dan ukuran lebih dari 12, kemudian mem-print 'Sedang' jika ruangan adalah 'Kamar' dan ukuran lebih dari 6 dan memprint 'Kecil' jika ruangan adalah 'Kamar' dan ukuran ruangan lebih kecil dan sama dengan 6.

```python
ukuran = int(input('Silahkan masukkan ukuran Kamar anda: '))

if ukuran > 12:
  print('Kamar anda Besar')
elif ukuran > 6:
  print('Kamar anda Sedang')
else:
  print('Kamar anda Kecil')
```

<h2>Soal 4: If-Elif-Else Statement</h2>

lengkapi baris kode dibawah dengan conditonal statement untuk mengkategorikan usia seseorang sesuai dengan kriteria berikut:
 - usia dibawah 5: balita
 - usia dibawah 13: anak-anak
 - usia dibawah 18: remaja
 - usia dibawah 65: dewasa
 - usia diatas 65: lanjut usia

output yang diminta (String): Termasuk Remaja

```python
usia = int(input('Masukkan Usia anda: '))

if usia <= 5:
  print('Balita')
elif usia <= 13:
  print('Anak-anak')
elif usia <= 18:
  print('Remaja')
elif usia <= 65:
  print('Dewasa')
else:
  print('Lanjut usia')
```

<h2>Soal 5: Program Interaktif If-Else Statement</h2>

Buat program interaktif yang akan mengeluarkan output "Kelipatan 7" apabila input yang dimasukkan adalah bilangan merupakan kelipatan 7, dan "Bukan Kelipatan 7" jika sebaliknya

```python
kelipatan = int(input('Masukkan Angka : '))

if kelipatan % 7 == 0:
  print('Kelipatan 7')
else:
  print('bukan kelipatan 7')
```

output yang diminta ketika input user angka kelipatan 7, maka akan mereturn string "Kelipatan 7". <br>Jika bukan kelipatan 7, maka akan mereturn string "Bukan Kelipatan 7

<h2>Soal 6: Program Interaktif 2 If-Else Statement</h2>

buatlah program interaktif untuk mengecek inputan user dengan ketentuan sebagai berikut:
<br>
-program akan menampilkan 'Input tidak Valid' jika yang dimasukkan bukan angka
<br>
-program akan menampilkan 'Input Diluar Range' jika angka yang dimasukkan tidak diantara 0-99
<br>
-program akan menampilkan 'Input Valid' jika input berupa angka dan ada di antara 0-99

```python
user = input('Masukkan Angka: ')

try:
  if int(user) <= 99:
    print('Input Valid')
  else:
    print('Input di luar Range')
except ValueError:
  print('Input tidak Valid')
```
