# Jawaban Nomor 1

```python
data = {
    'Tipe Ujian':['UTS1', 'UAS1'] * 3,
    'Mata Pelajaran':['Statistika', 'Fisika', 'Fisika', 'Statistika', 'Fisika', 'Statistika'],
    'Jumlah Peserta':[14, 17, 12, 16, 14, 13]
}

data = pd.DataFrame(data).set_index('Tipe Ujian')
data
```

# Jawaban Nomor 2

```python
df = {
    'Hari':['Sabtu', 'Minggu'],
    'Kebun Binatang':[271, 399],
    'Dufan':[501, 700],
    'Disney':[1000, 1001],
    'Bali':[900, 803]
}

df = pd.DataFrame(df).set_index('Hari')
df
```

```python
df.reset_index().melt(id_vars='Hari', var_name='Tempat Hiburan',
                      value_name='Pengunjung', value_vars=['Kebun Binatang', 'Dufan', 'Disney', 'Bali'])
```

# Jawaban Nomor 3

```python
df_diamond = pd.read_csv('/content/sample_data/diamonds-1.csv')

df_diamond
```

```python
def min_max(x):
  return x.max() - x.min()

df_diamond.groupby('cut')[['price', 'carat']].agg({'price':min_max, 'carat':'std'})
```
