## Arithmetic Function

## Arithmetic Function with Parameters

```
def perkalian(a, b):
    print('Total is :', a * b)

perkalian(a=40, b=30)
```

## Arithmetic Function with Parameters and Output

```
def pembagian(a, b, c=2):
    total = a * b / c
    return total

result = pembagian(10, 5)
print(result)
```

## Function with Conditional Statement

```
def wealth_estimator(money):
    if money >= 1000000:
        status = 'You are above 1%'

    elif money < 1000000 and money <= 100:
        status = 'You are in middle class'

    else:
        status = 'You are poor'

    return status

status = wealth_estimator(2000)
print(status)
```

## Function with Loop

```
numbers = [1, 2, 3, 4, 5]

def add_number(array, extra=2):
    placeholder = []

    for number in array:
        new_number = number + extra
        placeholder.append(new_number)

    return placeholder

result = add_number(numbers, extra=10)
```

## Function inside Function

```
def perkalian(a, b):
    return a * b

def multiply_number(array, extra=2):
    placeholder = []

    for number in array:
        new_number = perkalian(number, extra)
        placeholder.append(new_number)

    return placeholder

numbers = [1, 2, 3, 4, 5]
result = multiply_number(numbers, extra=10)
```

## Lower case, Upper case, Capitalize

```
# Lower Case
message7 = "HELLO WORLD"
message7a = message7.lower()
print(message7a)

# Upper Case
message7a = message7.upper()
print(message7a)

# Capitalize
message7a = message7.capitalize()
print(message7a)
```

## Replace

```
message8 = "HELLO WORLD"
message8a = message8.replace("L", "pizza")
print(message8a)
```
