```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('HR_comma_sep.csv')
```

```python
df.head()
```

# Barchart

```python
salary_count = df['salary'].value_counts()
```

```python
plt.figure(figsize = (10,7))
plt.bar(salary_count.index,salary_count)
plt.show()
```

# Histogram

```python
plt.figure(figsize=(10,7))
plt.hist(df['average_montly_hours'])
plt.show()
```

# Boxplot

```python
plt.figure(figsize=(7,10))
sns.boxplot(x = 'left',y='satisfaction_level',data = df)
plt.show()
```

# Violin Plot

```python
plt.figure(figsize=(7,10))
sns.violinplot(x = 'left',y='satisfaction_level',data = df)
plt.show()
```

# Scatter Plot

```python
df2 = pd.read_csv('Iris.csv')
```

```python
df2.head()
```

```python
plt.figure(figsize = (10,7))
sns.scatterplot(x = 'PetalLengthCm', y='PetalWidthCm',hue='Species',data = df2)
plt.show()
```

# Heatmap Plot

```python
HR_corr = df.corr()
```

```python
plt.figure(figsize = (10,8))
sns.heatmap(HR_corr,annot=True)
plt.show()
```
