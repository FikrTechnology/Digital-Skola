# Import Libraries

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
dataset = pd.read_csv('social_network_ads.csv')
dataset.head()
```

# Train Test Split

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X = pd.get_dummies(X)
X = X.drop('User ID', axis = 1)
X
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

 # Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train)
print(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
X_train
```

```python
X_test
```

# Training The Model

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5)
classifier.fit(X_train, y_train)
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
y_test_pred
```

```python
prediction = pd.DataFrame({'actual': y_test, 'predicted': y_test_pred})
prediction
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_test_pred)
```

```python
y_train_pred = classifier.predict(X_train)
accuracy_score(y_train, y_train_pred)
```

```python
print(classifier.predict(sc.transform([[30,87000, 1, 0]])))
```

# Search Best Number of K

```python
def search_k(X_train, X_test, y_train, y_test, maximum_k=5):
    acc_train = []
    acc_test = []
    ks = range(1, maximum_k+1)
    for k in ks:
        # Train based on number of K
        classifier = KNeighborsClassifier(n_neighbors=k)
        classifier.fit(X_train, y_train)
        
        # Predict the result
        y_train_pred = classifier.predict(X_train)
        y_test_pred = classifier.predict(X_test)
        
        # Training Performance
        score_train = accuracy_score(y_train, y_train_pred)
        acc_train.append(score_train)

        # Test Performance
        score_test = accuracy_score(y_test, y_test_pred)
        acc_test.append(score_test)

    result = pd.DataFrame({'k': ks, 'acc_train': acc_train, 'acc_test': acc_test})

    plt.figure(figsize=(12, 8))
    plt.title('KNN Performance')
    plt.ylabel('Accuracy')
    plt.xlabel('K')
    sns.lineplot(data=result, x='k', y='acc_train')
    sns.lineplot(data=result, x='k', y='acc_test')
    plt.grid()
    plt.xticks(result['k'])
    plt.show()
```

```python
search_k(X_train, X_test, y_train, y_test, maximum_k=15)
```

# Decision Tree

```python
dataset.head()
```

```python
# Separate Training and Test set
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

# Training Model

```python
from sklearn.tree import DecisionTreeClassifier
dectree = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth=5)
dectree.fit(X_train, y_train)
```

```python
y_test_pred = dectree.predict(X_test)
accuracy_score(y_test, y_test_pred)
```

```python
y_train_pred = dectree.predict(X_train)
accuracy_score(y_train, y_train_pred)
```

```python
print(dectree.predict([[30,87000]]))
```

# Search Best Tree

```python
def search_tree(X_train, X_test, y_train, y_test, max_depths):
    acc_train = []
    acc_test = []
    for depth in max_depths:
        # Train based on tree's depth
        classifier = DecisionTreeClassifier(criterion='entropy', random_state=0,
                                            max_depth=depth)
        classifier.fit(X_train, y_train)
        
        # Predict the result
        y_train_pred = classifier.predict(X_train)
        y_test_pred = classifier.predict(X_test)
        
        # Training Performance
        score_train = accuracy_score(y_train, y_train_pred)
        acc_train.append(score_train)

        # Test Performance
        score_test = accuracy_score(y_test, y_test_pred)
        acc_test.append(score_test)

    result = pd.DataFrame({'depths': max_depths, 'acc_train': acc_train, 'acc_test': acc_test})

    plt.figure(figsize=(12, 8))
    plt.title('Decision Tree Performance')
    plt.ylabel('Accuracy')
    plt.xlabel('Max Depths')
    sns.lineplot(data=result, x='depths', y='acc_train')
    sns.lineplot(data=result, x='depths', y='acc_test')
    plt.grid()
    plt.xticks(result['depths'])
    plt.show()
```

```python
search_tree(X_train, X_test, y_train, y_test, max_depths=[1,2,3,4,5,6,7,8,9])
```

# Visualize The Decision Tree

```python
from sklearn import tree

plt.figure(figsize=(25,12))
tree.plot_tree(dectree, feature_names=X.columns,
               class_names=['Not Buy', 'Buy'],
               filled=True)
plt.show()
```
