Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-white')

fig, ax = plt.subplots(figsize = (12, 8))

ax.plot(x, sin, color = 'r', label = 'Cosx')
ax.plot(x, cos, ls = '--', color = 'g', label = '3 Cosx')
ax.plot(x, sin2, ls = '-.', color = 'b', label = 'Sin (x - 90)')
ax.plot(x, cos3, ls = '', marker = 'x', color = 'y', label = '2 Sinx')

ax.set_xlabel('X-Label Axis')
ax.set_ylabel('Y-Label Axis')
ax.set_title('Grafik Cos Sin')
ax.legend(loc = 'center', fontsize = 'x-small')

plt.show()
```
