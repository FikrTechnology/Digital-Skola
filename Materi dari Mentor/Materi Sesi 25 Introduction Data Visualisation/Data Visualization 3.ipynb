```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
plt.style.use("seaborn-white")

x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(ncols = 3, sharey = False, figsize = (15,7))

ax[0].plot(x, tan, color = 'b', label = 'tanX') 
ax[1].plot(x, sin, color = 'r', label = 'sinX')
ax[2].plot(x, cos, color = 'g', label = 'cosX')
ax[0].set_title('Grafik tanX')
ax[1].set_title('Grafik sinX')
ax[2].set_title('Grafik cosX')
fig.legend(loc = "upper center")


plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
fig, ax = plt.subplots(figsize=(16,10))

ax.plot(x,y)
ax.set_title('Grafik CosX', fontsize='medium')

ins1 = ax.inset_axes([0.915,0.9,0.26,0.25])
ins1.plot(x,y4, color='black')
ins1.plot(x,y, color='b')
ins1.set_title('Grafik CosX^4', fontsize='medium')

ins2 = ax.inset_axes([-0.085,0.9,0.26,0.25])
ins2.plot(x,y5, color='y')
ins2.plot(x,y, color='b')
ins2.set_title('Grafik CosX^5', fontsize='medium')

ins3 = ax.inset_axes([-0.085,-0.1,0.26,0.25])
ins3.plot(x,y2, color='r')
ins3.plot(x,y, color='b')
ins3.set_title('Grafik CosX^2', fontsize='medium')

ins4 = ax.inset_axes([0.915,-0.1,0.26,0.25])
ins4.plot(x,y3, color='g')
ins4.plot(x,y, color='b')
ins4.set_title('Grafik CosX^3', fontsize='medium')

plt.show()
```
